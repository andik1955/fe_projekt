```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib

matplotlib.style.use('ggplot')
plt.rcParams['figure.figsize'] = 16, 12
```

```python
#drive = '/run/media/csaq7453/Elements/'
drive = 'E:/'
drive = drive + 'VU_Fernerkundung/Projekt/'

folderPixTest = drive + '20170627_1410UTC_sampleSizeTest/' + 'processingTime.txt'

folderCTest = drive + '20170627_0928UTC_svmParameterTest/' + 'processingTime.txt'

cTest = pd.read_csv(folderCTest, sep='\t', decimal='.')

pixTest = pd.read_csv(folderPixTest, sep='\t', decimal='.')
```

# C Parameter

### C und Dauer

```python
cTest.columns = ["C-Parameter", "Dauer um den Klassifikator zu trainieren", "Dauer des Klassifikationsprozesses", "Accuracy Score"]

cTest.plot(x = cTest["C-Parameter"], y = ["Dauer um den Klassifikator zu trainieren", "Dauer des Klassifikationsprozesses"], style=['b-','r-'])

plt.ylabel("Dauer [Sekunden]")
plt.xlabel("C-Parameter")
plt.title("Dauer des Prozesses in Abhaengigkeit vom C-Parameter")

ax = plt.gca()
ax.text(1.1, 0.1, '(2000 Trainingspixel)', transform=ax.transAxes, color='black', fontsize=11)


plt.show()
```

### C und Accuracy

```python
cTest.columns = ["C-Parameter", "Dauer um den Klassifikator zu trainieren", "Dauer der Klassifikationsprozesses", "Accuracy Score"]

cTest.plot(x = cTest["C-Parameter"], y = ["Accuracy Score"], style=['go'], legend=False)

plt.ylabel("Accuracy Score")
plt.xlabel("C-Parameter")
plt.ylim(0,1)
plt.xlim(0,6)
plt.title("Accuracy Score in Abhaengigkeit vom C-Parameter")

ax = plt.gca()
ax.text(1.1, 0.1, '(2000 Trainingspixel)', transform=ax.transAxes, color='black', fontsize=11)

plt.show()
```

# Sample Size

### Pixelnumber und Dauer

```python
pixTest.columns = ["Anzahl Trainingspixel", "Dauer um den Klassifikator zu trainieren", "Dauer des Klassifikationsprozesses", "Accuracy Score"]

pixTest.plot(x = pixTest["Anzahl Trainingspixel"], y = ["Dauer um den Klassifikator zu trainieren", "Dauer des Klassifikationsprozesses"], style=['b-','r-'])

plt.ylabel("Dauer [Sekunden]")
plt.xlabel("Anzahl der Trainingspixel")
plt.title("Dauer des Prozesses in Abhaengigkeit vom C-Parameter")

ax = plt.gca()
ax.text(0.05, -0.25, 'LinearSVC mit C = XX', transform=ax.transAxes, color='black', fontsize=11)


plt.show()
```

### Pixelnummer und Accuracyscore

```python
pixTest.columns = ["Anzahl Trainingspixel", "Dauer um den Klassifikator zu trainieren", "Dauer des Klassifikationsprozesses", "Accuracy Score"]

pixTest.plot(x = pixTest["Anzahl Trainingspixel"], y = ["Accuracy Score"], style=['go'])

plt.ylabel("Dauer [Sekunden]")
plt.xlabel("Anzahl der Trainingspixel")
plt.title("Dauer des Prozesses in Abhaengigkeit vom C-Parameter")

ax = plt.gca()
ax.text(0.05, -0.25, 'LinearSVC mit C = XX', transform=ax.transAxes, color='black', fontsize=11)


plt.show()
```
